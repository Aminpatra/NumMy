```python
def fact(n):
  
  if n == 1:
    return 1
  
  return n * fact(n-1)


print(fact(5))
```

```python
def arr_shape(List, shape = None):

  if shape is None:
    shape = []

  if not isinstance(List, list) or not List:
    return shape
  
  shape.append(len(List))
  
  return arr_shape(List[0], shape)

print(arr_shape(d1))
print(arr_shape(d2))
print(arr_shape(d3))
print(arr_shape(d_not))
```
